### ðŸ”¢ What Does K Mean?

- K represents the **number of clusters**
- If K = 2 â†’ two clusters
- If K = 3 â†’ three clusters, and so on

The value of K must be specified before running the algorithm.

### ðŸ“˜ Definition of K-Means

K-Means is an iterative algorithm that divides an unlabeled dataset into K different clusters such that each data point belongs to exactly one cluster with similar properties.

## ðŸŽ¯ Centroid-Based Clustering

K-Means is a **centroid-based algorithm**:
- Each cluster is represented by a centroid
- The centroid is the point with minimum cumulative distance
- The objective is to minimize the distance between data points and centroids

## ðŸ§­ Aim of the K-Means Algorithm

The K-Means algorithm performs two main tasks:

1. Determine the best K centroids using an iterative process
2. Assign each data point to the nearest centroid, forming clusters

## âš™ï¸ How K-Means Algorithm Works

The K-Means algorithm follows these steps:

1. Plot the data
2. Choose the number of clusters (K)
3. Select random K points as centroids
4. Assign each data point to the nearest centroid
5. Recalculate centroids
6. Repeat until clusters no longer change

# Introduction to K-Means Clustering

**K-Means clustering** is an unsupervised machine learning algorithm used to partition `n` observations into `k` clusters, where each observation belongs to the cluster with the nearest mean (centroid). It's an iterative algorithm that aims to minimize the within-cluster sum of squares (WCSS), also known as inertia.

**How it Works (Simple Example):**
1.  **Initialization:** Choose `k` random data points as initial centroids for the `k` clusters.
2.  **Assignment Step:** Assign each data point to the closest centroid, forming `k` clusters.
3.  **Update Step:** Recalculate the centroids as the mean of all data points assigned to that cluster.
4.  **Iteration:** Repeat steps 2 and 3 until the centroids no longer change significantly or a maximum number of iterations is reached.

**Common Use Cases:**
*   **Customer Segmentation:** Grouping customers based on purchasing behavior or demographics.
*   **Image Compression:** Reducing the number of colors in an image by grouping similar colors.
*   **Document Clustering:** Organizing large collections of text documents into topics.
*   **Anomaly Detection:** Identifying unusual patterns or outliers in data.
*   **Geospatial Analysis:** Grouping locations with similar characteristics.

**Pros of K-Means Clustering:**
*   **Simplicity:** Easy to understand and implement.
*   **Efficiency:** Relatively fast for large datasets, especially when `k` is small.
*   **Scalability:** Can handle a large number of samples.
*   **Versatility:** Applicable to various types of data and problems.

**Cons of K-Means Clustering:**
*   **Requires `k` in Advance:** The number of clusters (`k`) must be specified beforehand, which can be challenging to determine.
*   **Sensitive to Initial Centroids:** Different initial centroid choices can lead to different clustering results.
*   **Sensitive to Outliers:** Outliers can significantly affect cluster centroids and overall results.
*   **Assumes Spherical Clusters:** K-Means works best with spherically shaped clusters of similar size and density.
*   **Difficulty with Non-Globular Clusters:** Struggles with clusters of irregular shapes or varying densities.

## Import Libraries

This cell imports the necessary libraries, `numpy` for numerical operations and `pandas` for data manipulation.

```python
import numpy as np
import pandas as pd
```

## Load and Inspect Data

This cell loads the 'student_clustering.csv' file into a pandas DataFrame, prints its shape, and displays the first few rows.

```python
df = pd.read_csv('student_clustering.csv')
print("The shape of data is",df.shape)
df.head()
```

## Visualize Data Distribution

This cell imports `matplotlib` for plotting and generates a scatter plot to visualize the distribution of 'cgpa' and 'iq'.

```python
import matplotlib.pyplot as plt
plt.scatter(df['cgpa'],df['iq'])
```

## Import K-Means

This cell imports the `KMeans` class from the `sklearn.cluster` module, which is used for K-means clustering.

```python
from sklearn.cluster import KMeans
```

## Calculate WCSS for Elbow Method

This cell calculates the Within-Cluster Sum of Squares (WCSS) for different numbers of clusters (1 to 10). WCSS is used to determine the optimal number of clusters using the elbow method.

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

## Display WCSS Values

This cell displays the calculated WCSS values.

```python
wcss
```

## Plot Elbow Curve

This cell plots the WCSS values against the number of clusters to visualize the elbow curve, which helps in identifying the optimal number of clusters.

```python
plt.plot(range(1,11),wcss)
```

## Apply K-Means Clustering

This cell extracts the data into array `X`, initializes `KMeans` with 4 clusters (chosen from the elbow method), and applies the clustering algorithm to the data, storing the cluster assignments in `y_means`.

```python
X = df.iloc[:,:].values
km = KMeans(n_clusters=4)
y_means = km.fit_predict(X)
```

## Display Cluster Assignments

This cell displays the cluster assignments for each data point.

```python
y_means
```

## Access Cluster Data (Example)

This cell demonstrates how to access specific data points belonging to a particular cluster (in this case, the second feature (index 1) of data points assigned to cluster 3).

```python
X[y_means == 3,1]
```

## Visualize Clustered Data

This cell generates a scatter plot to visualize the clustered data, with each cluster represented by a different color.

```python
plt.scatter(X[y_means == 0,0],X[y_means == 0,1],color='blue')
plt.scatter(X[y_means == 1,0],X[y_means == 1,1],color='red')
plt.scatter(X[y_means == 2,0],X[y_means == 2,1],color='green')
plt.scatter(X[y_means == 3,0],X[y_means == 3,1],color='yellow')
```

## Generate 3D Sample Data

This cell imports `make_blobs` to generate a 3-dimensional dataset with specified centroids and cluster standard deviations for K-means clustering demonstration.

# K-Means on 3-D Data

```python
from sklearn.datasets import make_blobs

centroids = [(-5,-5,5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]
cluster_std = [1,1,1,1]

X,y = make_blobs(n_samples=200,cluster_std=cluster_std,centers=centroids,n_features=3,random_state=1)
```

## Display 3D Sample Data

This cell displays the generated 3D sample data.

```python
X
```

## Visualize 3D Data (Unclustered)

This cell imports `plotly.express` and creates an interactive 3D scatter plot of the generated data before clustering.

```python
import plotly.express as px
fig = px.scatter_3d(x=X[:,0], y=X[:,1], z=X[:,2])
fig.show()
```

## Calculate WCSS for 3D Data

This cell calculates the WCSS for the 3D dataset for a range of clusters (1 to 20) to prepare for the elbow method.

```python
wcss = []
for i in range(1,21):
    km = KMeans(n_clusters=i)
    km.fit_predict(X)
    wcss.append(km.inertia_)
```

## Plot Elbow Curve for 3D Data

This cell plots the WCSS values for the 3D data against the number of clusters to find the optimal number of clusters using the elbow method.

```python
plt.plot(range(1,21),wcss)
```

## Apply K-Means to 3D Data

This cell applies K-Means clustering with 4 clusters to the generated 3D data and stores the cluster predictions in `y_pred`.

```python
km = KMeans(n_clusters=4)
y_pred = km.fit_predict(X)
```

## Create DataFrame with Cluster Labels

This cell creates a pandas DataFrame from the 3D data and adds a new column 'label' to store the cluster assignments.

```python
df = pd.DataFrame()

df['col1'] = X[:,0]
df['col2'] = X[:,1]
df['col3'] = X[:,2]
df['label'] = y_pred
```

## Visualize Clustered 3D Data

This cell creates and displays an interactive 3D scatter plot of the clustered data, color-coded by the assigned labels.

```python
fig = px.scatter_3d(df,x='col1', y='col2', z='col3',color='label')
fig.show()
```
